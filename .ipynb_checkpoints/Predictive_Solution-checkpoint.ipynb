Using Recursive Feature Elimination method which uses the model accuracy to identify which attributes contribute the most to predicting the target attribute.

Import Python Data Analysis Library to extract the features.

```python
import pandas as pd
```

Import Python Numpy Library to hold a N-dimensional array.

```python
import numpy as np
```

Read CSV as df to analysis the Dataset.

```python
df=pd.read_csv('../Task-for-CeleraOne/DataSet/task_data.csv')
```

Employing Pandas and Numpy to extract names of each Column and Features and Lables.

```python
sensorsName = np.array(list(df.columns.values))[2:]
sensorReadings = df.as_matrix()[:,2:]
lable = df['class_label'].as_matrix()
```

Select the features with Scikit-Learn Laibrary.

I am going to rank the predicting the target attribute by a feature selection approach such as Recursive Feature Elimination.

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

Create a classifier used to evaluate a subset of attributes. Also create the RFE model and select 2 attributes.

```python
rfe = RFE(LogisticRegression(), 2)
rfe = rfe.fit(sensorReadings, lable)
```

Sort the ranking where the first sensor is the most predictable one in Recursive Feature Elimination. 

```python
rankByRFE = sorted(zip(map(lambda x: round(x, 2), rfe.ranking_), sensorsName), reverse=True)
```

```python
rankByRFEforVisualize = (rfe.ranking_)
```

Here I will show a Visualization of the method's result. 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(12, 5))
plt.xticks(np.arange(len(list(zip(sensorsName))))) 
plt.plot(list(zip(rankByRFEforVisualize)))
plt.xlabel('List of Sensors')
plt.ylabel('Score of Sensors')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.bar(np.arange(len(list(zip(*rankByRFE))[1])), list(zip(*rankByRFE))[0], align='center')
plt.xticks(np.arange(len(list(zip(*rankByRFE))[1])), list(zip(*rankByRFE))[1]) 

plt.xlabel('List of Sensors')
plt.ylabel('Score of Sensors')
plt.show()
```

Save the list to a CSV file.

```python
import csv
with open('../Task-for-CeleraOne/DataSet/rank_by_RecursiveFeatureElimination.csv', 'w', newline='') as out_RFEfile:
    a = csv.writer(out_RFEfile)
    data= rankByRFE
    a.writerows(data)
```
